```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def omegaDot_B(state, t):
    eulerRate = 20*np.pi/180 * np.array([np.sin(0.1*t), 0.01, np.cos(0.1*t)])
    return eulerRate

initialState = np.radians([40, 30, 80])
tEnd = 42
step = 0.01
t = np.arange(0, tEnd + step, step)
sol = odeint(omegaDot_B, initialState, t)

norm = np.linalg.norm(sol[-1, :])
print(f"Norm of angle at time step of {tEnd} is {norm}")
plt.figure(1)
plt.plot(t, sol, linewidth=2)
plt.legend(['$\\psi$', '$\\theta$', '$\\phi$'], loc='best')
plt.grid(True)
plt.show()
```
